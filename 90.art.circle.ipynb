```python
#RANDOM CIRCLE ART

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def determine_canvas_size():
    height_size = np.random.choice([6, 9, 12]) * (4 - density)              #canvas y size
    width_size = np.random.choice([8, 9, 10]) * (4 - density)               #canvas x size
    
    return(np.array([height_size, width_size]))                             #return canvas size as length 2 list
```

```python
def establish_circles():
    (x_vals, y_vals, sizes) = (np.empty(quantity), np.empty(quantity), np.empty(quantity))
    clrs = []
    color_choice = np.random.choice(['w', 'k', 'y', 'b', 'r', 'g'], np.random.randint(1, 4))
    
    for i in range(quantity):                                                 #establish the following criteria for circles
        x_vals[i] = np.random.uniform() * aspect[1]                           #create array of x values
        y_vals[i] = np.random.uniform() * aspect[0]                           #create array of y values for 'pairs'
        sizes[i] = np.random.choice([1, 3, 6, 9, 12, 15]) * circle_size * 750 #create array of sizes
        clrs.append(np.random.choice(color_choice))                           #create list of colors
        
    return(x_vals, y_vals, sizes, clrs)
```

```python
####################################################################### MAIN MAIN
print(); print('YOU choose the following parameters to create random CIRCLE ART'); print(); 

#establish figure, axes objects
fig, ax = plt.subplots(2, 2, figsize = (20, 25))
plt.subplots_adjust(wspace = 0.35)

density = int(input('Select Density Level:  High (3), Medium (2), Low (1)'))           #density parameter input
circle_size = int(input('Select Circle Size:  Large (3), Small (1)'));                 #circle size parameter input
back_color = input('Select a Background Color:  White (w), Black (k), Yellow (y), Blue(b), Red (r), Green (g)'); print(); print()

circle_outline = input('  Include Circle Outlines:  Yes (y), No (n)')                  #determine if circles outlined and color if so
if circle_outline == 'y':
    outline_color = np.random.choice(['w', 'k', 'y', 'b', 'r', 'g'])
else:
    outline_color = 'None'
    
transp = float(input('Select Transparency Level:  High (0.25), Med (0.5), None (1.0)')) #establish tranparency level (alpha)

for i in range(2):                                                                     #establish (2 stage) loop for subplots
    for j in range(2):                                               
        aspect = determine_canvas_size()                                               #determine size of 'canvas' height by width
        quantity = np.random.choice([2, 4, 6]) * (density * 4)                         #determine quantity of circles
        (x_loc, y_loc, sizes, clrs) = establish_circles()                              #create 1d arrays of x, y, size
        #create plots
        ax[i, j].tick_params(bottom = False, left = False, labelbottom = False, labelleft = False)
        ax[i, j].set(facecolor = back_color, xlim = (0, aspect[1]), ylim = (0, aspect[0]))
        ax[i, j].scatter(x_loc, y_loc, s = sizes, c = clrs, edgecolors = outline_color, alpha = transp)
```
